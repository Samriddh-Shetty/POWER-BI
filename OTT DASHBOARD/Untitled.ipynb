```python
import pandas as pd
```

```python
netflix=pd.read_csv("netflix_titles.csv")
```

```python
netflix
```

```python
netflix.isnull().sum()/len(netflix)*100
```

```python
netflix['director']=netflix['director'].fillna("Not Given")
```

```python
netflix
```

```python
netflix.dropna(subset=['cast','country','rating','duration'],inplace=True)
```

```python
netflix
```

```python
netflix.drop(['date_added',"show_id"],axis=1,inplace=True)
```

```python
netflix
```

```python
netflix.isnull().sum()/len(netflix)*100
```

```python
netflix.to_csv('netflix.csv', index=False)
```
